```python
from google.colab.patches import cv2_imshow
import cv2

# Load the image
image = cv2.imread("product1.jpg")

# Convert the image to grayscale
grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the original image
cv2_imshow(image)

# Apply Canny edge detection
canny = cv2.Canny(grayscale, 100, 150)

# Find contours
contours, hierarchy = cv2.findContours(canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Print the number of contours
print("Number of contours =", len(contours))

# Display the Canny edge-detected image
cv2_imshow(canny)

# Draw contours on the original image
cv2.drawContours(image, contours, -1, (0, 255, 0), 3)

# Display the image with contours drawn
cv2_imshow(image)
```

```python
for i, contour in enumerate(contours):
  area = cv2.contourArea(contour)
  perimeter = cv2.arcLength(contour,True)
  print(f"Contour {i+1}: Area = {area}, Perimeter = {perimeter}")
  epsilon = 0.02*cv2.arcLength(contour,True)
  approx = cv2.approxPolyDP(contour,epsilon,True)
```

```python
for contour in contours:
  x,y,w,h = cv2.boundingRect(contour)
  cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
cv2_imshow(image)
```

```python
from google.colab.patches import cv2_imshow
import cv2

# Load the image
image = cv2.imread("product1.jpg")

# Set height and width (your known size)
h, w = 474, 474

# Compute the center of the image
center = (w // 2, h // 2)

# Generate the rotation matrix (90 degrees counter-clockwise)
M = cv2.getRotationMatrix2D(center, 270, 1.0)

# Perform the rotation using your dimensions
rotated = cv2.warpAffine(image, M, (w, h))

# Display the rotated image
cv2_imshow(rotated)
```

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Load the image
image = cv2.imread("product1.jpg")

# Translation values: shift 100 right and 50 down
tx = 100  # shift along x-axis
ty = 50   # shift along y-axis

# Create the translation matrix
M = np.float32([[1, 0, tx], [0, 1, ty]])

# Apply the translation
translated = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))

# Display the translated image
cv2_imshow(translated)
```
